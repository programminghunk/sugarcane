```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#!pip install scikit-learn

# Step 1: Read data from CSV file
file_path_v3 = 'results_v3.csv'  # Update with your CSV file path
file_path_v5 = 'results_v5.csv'  # Update with your CSV file path
file_path_v6 = 'results_v6.csv'  # Update with your CSV file path
file_path_v7 = 'results_v7.csv'  # Update with your CSV file path
file_path_v8 = 'results_v8.csv'  # Update with your CSV file path
file_path_v9 = 'results_v9.csv'  # Update with your CSV file path
file_path_v10= 'results_v10.csv'
file_path_v8_custom = 'results_v8_custom.csv'  # Update with your CSV file path
data_v3 = pd.read_csv(file_path_v3)
data_v5 = pd.read_csv(file_path_v5)
data_v6 = pd.read_csv(file_path_v6)
data_v7 = pd.read_csv(file_path_v7)
data_v8 = pd.read_csv(file_path_v8)
data_v9 = pd.read_csv(file_path_v9)
data_v10 = pd.read_csv(file_path_v10)
data_v8_custom = pd.read_csv(file_path_v8_custom)
column_0 = data_v3.iloc[:, 0]
column_1 = data_v3.iloc[:, 1]
column_2 = data_v5.iloc[:, 1]
column_3 = data_v6.iloc[:, 1]
column_4 = data_v7.iloc[:, 1]
column_5 = data_v8.iloc[:, 1]
column_6 = data_v9.iloc[:, 1]
column_7 = data_v10.iloc[:,1]
column_8 = data_v8_custom.iloc[:, 1]

# Increase figure size for better readability
plt.figure(figsize=(12, 6))
palette = sns.color_palette("husl", 10)
sns.lineplot(x=column_0, y=column_1,label='yolo_v3_tiny',color=palette[0],linestyle="-")
sns.lineplot(x=column_0, y=column_2,label='yolo_v5_nano',color=palette[1],linestyle="--")
sns.lineplot(x=column_0, y=column_3,label='yolo_v6_nano',color=palette[2],linestyle="-.")
sns.lineplot(x=column_0, y=column_4,label='yolo_v7_tiny',color=palette[3],linestyle=":")
sns.lineplot(x=column_0, y=column_5,label='yolo_v8_nano',color=palette[4], linestyle="-")
sns.lineplot(x=column_0, y=column_6,label='yolo_v9_tiny',color=palette[5], linestyle="--")
sns.lineplot(x=column_0, y=column_7,label='yolo_v10_nano',color=palette[7], linestyle="-.")
sns.lineplot(x=column_0, y=column_8,label='yolo_v8_tinier(ours)', color='black',linestyle="-")
#plt.title('mAP50 Comparison of YOLO Tiny Models (v3 to v10)', fontsize=16)  # Increase title font size
plt.title('Box Loss(train): Comparison of YOLO Tiny Models (v3 to v10)', fontsize=16)
plt.xlabel('Epochs', fontsize=12)  # Increase x-axis label font size
plt.ylabel('mAP50', fontsize=12)  # Increase y-axis label font size

# Customize legend for better visibility
plt.legend(fontsize=10, loc='upper right')  # Adjust legend font size and location

# Add grid lines for better visual structure
plt.grid(True)

plt.show()
```
