```python
from PIL import Image

# Load the images
img1 = Image.open("recall.png")
img2 = Image.open("precision.png")
img3 = Image.open("map50.png")
img4 = Image.open("map95.png")

# Resize all images to the same size (optional but recommended)
size = (img1.width, img1.height)
img2 = img2.resize(size)
img3 = img3.resize(size)
img4 = img4.resize(size)

# Create a new blank image for the collage
collage_width = size[0] * 2
collage_height = size[1] * 2
collage = Image.new("RGB", (collage_width, collage_height), color=(255, 255, 255))

# Paste the images into a 2x2 grid
collage.paste(img1, (0, 0))                        # Top-left
collage.paste(img2, (size[0], 0))                  # Top-right
collage.paste(img3, (0, size[1]))                  # Bottom-left
collage.paste(img4, (size[0], size[1]))            # Bottom-right

# Save the collage
collage.save("collage_2x2.png")
print("Collage saved as collage_2x2.png")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

# Use Times New Roman for academic look
mpl.rcParams['font.family'] = 'Times New Roman'
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 12

# Data for models
models = ['Proposed Model', 'YOLOv8n', 'YOLOv9n']
map50_95 = [0.624, 0.6254, 0.6152]
tensorrt_fps = [386.10, 260.51, 219.14]
jetson_nano_fps = [47, 35, 40]  # Approx average
model_size = [4.6, 6, 4.4]

# Create larger figure for better clarity
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Model Comparison: Accuracy, Speed, and Size', fontsize=18)

# Accuracy
sns.barplot(x=models, y=map50_95, ax=axs[0, 0], palette="summer")
axs[0, 0].set_title("Accuracy (mAP50-95)")
axs[0, 0].set_ylabel("mAP")

# TensorRT FPS
sns.barplot(x=models, y=tensorrt_fps, ax=axs[0, 1], palette="Blues")
axs[0, 1].set_title("Inference Speed (TensorRT)")
axs[0, 1].set_ylabel("FPS")

# Jetson Nano FPS
sns.barplot(x=models, y=jetson_nano_fps, ax=axs[1, 0], palette="Oranges")
axs[1, 0].set_title("Jetson Nano Inference FPS")
axs[1, 0].set_ylabel("FPS")

# Model Size
sns.barplot(x=models, y=model_size, ax=axs[1, 1], palette="Greys")
axs[1, 1].set_title("Model Size (PyTorch)")
axs[1, 1].set_ylabel("Size (MB)")

# Tight layout and high resolution saving
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("high_quality_model_comparison.png", dpi=600)
plt.show()
```
